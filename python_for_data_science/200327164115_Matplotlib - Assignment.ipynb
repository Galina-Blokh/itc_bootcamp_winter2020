# Question #4

Create a pie-chart using the given data, to show how many people like each activity. Add a legend and an appropriate title. Locate the legend and the title so that they are not on top of the plot.

**Food for thought:** Can you read the plot without looking at the data? Can you see if the section "play" is larger than "eat"? What is the ratio between "work" and "study"? These are questions we would like to answer using plots

```python
activities = ['eat', 'sleep', 'work', 'play', 'study'] 
  
slices = [15, 30, 8, 16, 35] 
  
colors = ['r', 'y', 'g', 'b', 'c'] 

# Pie chart, where the slices will be ordered and plotted counter-clockwise:

explode = (0, 0, 0, 0, 0.1)  # only "explode" the 5th slice (i.e. 'study')

fig1, ax1 = plt.subplots()
ax1.pie(slices, explode=explode,  shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True,labels=activities, ncol=5)
plt.text(-1.1, -1.8, 'Question #4 how many people like each activity.')
plt.show()
```

# Question #6

Create four subplots in a plot, and fill each plot with a different function of your choice. color each plot in a different color.

```python
df = pd.DataFrame({'x': range(1,11), \
                   'y_1': np.random.randn(10), \
                   'y_2': np.random.randn(10)+range(1,11), \
                   'y_3': np.random.randn(10)+range(11,21),\
                   'y_4': np.random.randn(10)+range(41,51)})

fig, axs = plt.subplots(2, 2, gridspec_kw={'hspace': 0.5})

axs[0, 0].plot(df.x, df.y_1, 'tab:pink')
axs[0, 0].set_title('y_1')
axs[0, 1].plot(df.x, df.y_2, 'tab:orange')
axs[0, 1].set_title(' y_2')
axs[1, 0].plot(df.x, df.y_3, 'tab:green')
axs[1, 0].set_title(' y_3')
axs[1, 1].plot(df.x, df.y_4, 'tab:red')
axs[1, 1].set_title(' y_4')

for ax in axs.flat:
    fig.suptitle('Question #6 plot of 4  different functions')

plt.show()
```

# Question #7

Create the same subplots as in Question #6, but now have the plots sharing their x and y axis.

```python
df = pd.DataFrame({'x': range(1,11), \
                   'y_1': np.random.randn(10), \
                   'y_2': np.random.randn(10) + range(1,11), \
                   'y_3': np.random.randn(10)+range(11,21),\
                   'y_4': np.random.randn(10)+range(41,51)})

# i have 4 different functions, so they cant share y-axes
fig, axs = plt.subplots(4,sharex=True)

y1, = axs[0].plot(df.x, df.y_1, 'tab:pink')

y2, = axs[1].plot(df.x, df.y_2, 'tab:orange')

y3, = axs[2].plot(df.x, df.y_3, 'tab:green')

y4, = axs[3].plot(df.x, df.y_4, 'tab:red')


plt.legend([y1,y2,y3,y4], ['y1','y2','y3','y4'],loc='upper center', bbox_to_anchor=(0.5,-0.5),fancybox=True, shadow=True, ncol=4).get_frame()

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    fig.suptitle('Question #7 plot of 4 different functions sharing their x axis.')
    
```

# Question #9

Read the data from https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv into a pandas dataframe.
 1. Drop all rows with empty values in the columns "RPM", "Type" or "Price"
 2. Plot the relation between the variables "RPM" and "Price" on a scatter plot
 3. Color the data points by the variable "Type"
 4. Add a title and axis names
 5. Add a legend

```python
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
data
```

```python
# drop all rows with empty values in the columns "RPM", "Type" or "Price"
data_modif = data.dropna( how='any',subset=['RPM',"Type","Price"])
data_modif.head()
```

```python

# plot the relation between the variables "RPM" and "Price" on a scatter plot
f = plt.figure()
plt.scatter(data_modif.Price,data_modif.RPM)

plt.colorbar()
plt.show()
```

```python
# Color the data points by the variable "Type"

f, ax = plt.subplots()
type_groups = data_modif.groupby("Type")

for type_name, group in type_groups:
    ax.scatter(group.Price, group.RPM, label=type_name, alpha=0.35)
    

plt.show()
```

```python
# Add a title and axis names
f, ax = plt.subplots()
type_groups = data_modif.groupby("Type")

for type_name, group in type_groups:
    ax.scatter(group.Price, group.RPM, label=type_name, alpha=0.35)
    
plt.ylabel('RPM')
plt.xlabel('Price')
plt.title('Question #9: RPM, Price group by Type')

plt.show()
```

```python
# Add a legend
f, ax = plt.subplots()
type_groups = data_modif.groupby("Type")

for type_name, group in type_groups:
    ax.scatter(group.Price, group.RPM, label = type_name, alpha=0.35)
    
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0.1,title = "Car Type")
plt.ylabel('RPM')
plt.xlabel('Price')
plt.title('Question #9: RPM, Price group by Type')

plt.show()
```
